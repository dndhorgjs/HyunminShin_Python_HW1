```python
class Node():
    def __init__(self):
        self.data=None
        self.link=None

def printNodes(start):
    current = start
    if current==None:
        return
    print(current.data,end=' ')
    while current.link !=None:
        current=current.link
        print(current.data,end=' ')
    print()
    
def makeSimpleLinkedList(TwiceNumber):
    global memory,head,current,pre
    printNodes(head)
    node=Node()
    node.data=TwiceNumber
    memory.append(node.data)
    if head==None:
        head=node
        return
    
    if head.data[1]<TwiceNumber[1]:
        node.link=head
        head=node
        return
    
    current=head
    while current.link != None:
        pre=current
        current=current.link
        if current.data[1]<TwiceNumber[1]:
            node.link=current
            pre.link=current
            return
    
    current.link=node
    
def deleteNode(deleteData):
    global memory,head,current,pre
    
    if head.data==deleteData:
        current=head
        head=head.link
        del(current)
        return
    
    current=head
    while current.link != None:
        pre=current
        current=current.link
        if current.data==deleteData:
            pre.link=current.link
            del(current)
            return
        
def delete_data(position):
    if position<0 or position>len(memory)-1:
        print("   .")
        return
        
        
memory=[]
head,current,pre=None,None,None
select=-1

if __name__=="__main__":
    while (select!=3):
        select=int(input("(1: , 2: , 3:)--> "))
        if select==1:
            data=input(" -->")
            makeSimpleLinkedList(data)
            print(memory)
        elif select==2:
            pos=int(input(" -->"))
            if pos<0 or pos>len(memory)-1:
                print("   .")
    
            deleteNode(memory[pos-1])
            printNodes(head)
        elif select==3:
            print()
        else:
            print("1~3   ")
```
